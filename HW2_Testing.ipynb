# Networkx Tests

```python
import networkx as nx
import matplotlib.pyplot as plt
from search import graph
```

```python
network_file = './data/tiny_network.adjlist'
tiny_network = nx.read_adjlist('./data/tiny_network.adjlist', create_using=nx.DiGraph, delimiter=";")
tiny_network.nodes()
```

```python
list(tiny_network.neighbors('31806696'))
```

```python
nx.is_directed(tiny_network)
```

```python
nx.is_connected(tiny_network)
```

```python
nx.is_weakly_connected(tiny_network)
```

```python
nx.draw(tiny_network,with_labels=True, font_weight='bold') # Creates visual of node network.
plt.show()
```

```python
citation_file = './data/citation_network.adjlist'
citation_network = nx.read_adjlist('./data/citation_network.adjlist', create_using=nx.DiGraph, delimiter=";")
citation_network.nodes()
```

```python
list(citation_network.neighbors('31806696'))
```

```python
nx.is_directed(citation_network)
```

```python
nx.shortest_path(citation_network, 'Lani Wu', 'Nevan Krogan')
```

```python
for node in list(nx.nodes(citation_network)):
    if node not in list(nx.bfs_tree(citation_network, "Luke Gilbert")):
        print(node)
```
